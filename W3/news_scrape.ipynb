```python
from newspaper import Article

# List of news article URLs you want to download
article_urls = [
    'https://indianexpress.com/article/technology/opinion-technology/screenless-wearables-whoop-oura-ring-discreet-fitness-tracking-10368687/',

    'https://www.thehindu.com/sci-tech/technology/what-are-the-digital-personal-data-protection-rules-and-when-do-they-apply/article70283392.ece',
    'https://www.indiatoday.in/technology/features/story/oneplus-15-review-in-9-points-big-performance-bigger-battery-but-is-that-enough-2819702-2025-11-14',
    'https://www.indiatoday.in/technology/news/story/iphone-18-vs-samsung-galaxy-s26-both-phones-are-set-to-launch-next-year-here-are-rumoured-details-2820760-2025-11-16',
    'https://www.wired.com/story/all-my-employees-are-ai-agents-so-are-my-executives/',
    'https://www.computerweekly.com/news/366634578/Cl0p-claims-ransomware-hit-on-NHS',
    'https://apnews.com/article/trump-economy-midterms-republicans-affordability-election-losses-0e15f928cdbb01eee41eadfb477cb981',
    'https://apnews.com/article/haiti-embassy-attack-gangs-7c3b602a5ea0b1211a8e5c29c9096e7d'
    # Add 4-9 more URLs here
]

for i, url in enumerate(article_urls):
    try:
        article = Article(url)
        article.download()
        article.parse()

        # The article.text variable contains the clean, extracted body text
        clean_text = article.text

        # Create a filename (e.g., 'article_1.txt', 'article_2.txt', etc.)
        filename = f"article_{i+1}.txt"

        # Save the text to a separate .txt file
        with open(filename, 'w', encoding='utf-8') as f:
            f.write(clean_text)

        print(f"Successfully downloaded and saved: {filename}")

    except Exception as e:
        print(f"Could not process URL {url}. Error: {e}")
        continue
```
